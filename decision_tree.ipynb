```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics
import seaborn as sn
```

```python
col_names = ['pregnant', 'glucose', 'bp', 'skin', 'insulin', 'bmi', 'pedigree', 'age', 'label']
data = pd.read_csv('/content/diabetes.csv', header=0, names=col_names)
print(data.shape)
data.head()
```

```python
data.isnull().sum()
```

```python
feature_cols = ['pregnant', 'insulin', 'bmi',
                'age', 'glucose', 'bp', 'pedigree']

x = data[feature_cols]
y = data.label

# splitting the dataset into Training and Testing Dataset
x_train, x_test, y_train, y_test = train_test_split(x, y,
                                                    test_size=0.2, random_state=5)

display(x_train.shape, y_train.shape, x_test.shape, y_test.shape)
```

```python
model = DecisionTreeClassifier(criterion='entropy', random_state=5)
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
print('y_pred: ', y_pred)
```

```python
conf_mat = metrics.confusion_matrix(y_test, y_pred)
print('Confusion Matrix : ', conf_mat)
Accuracy_score = metrics.accuracy_score(y_test, y_pred)
print('Accuracy Score : ', Accuracy_score)
```

Predicting House Purchase Based on Age and Income

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
import matplotlib.pyplot as plt

# Sample dataset
data = {
    'Age': [25, 35, 40, 29, 50, 22, 33, 45],
    'Income': [40000, 80000, 100000, 50000, 120000, 30000, 75000, 110000],
    'Bought_House': [0, 1, 1, 0, 1, 0, 1, 1]  # 1 = bought, 0 = didn’t buy
}

df = pd.DataFrame(data)

# Features and label
X = df[['Age', 'Income']]
y = df['Bought_House']

# Model
model = DecisionTreeClassifier(criterion='entropy', random_state=0)
model.fit(X, y)


# Plot the tree
plt.figure(figsize=(10, 6))
tree.plot_tree(model, feature_names=['Age', 'Income'], class_names=['No', 'Yes'], filled=True)
plt.show()
```

```python
tree.plot_tree(model,
               feature_names=['Age', 'Income'],
               class_names=['No', 'Yes'],
               filled=True,
               rounded=True,
               fontsize=12)
```

Classify Activity Based on Weather (Go Out / Stay In)

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
import matplotlib.pyplot as plt

# Sample dataset
data = {
    'Weather': ['Sunny', 'Sunny', 'Rainy', 'Rainy', 'Sunny', 'Rainy'],
    'Activity': ['Go Out', 'Go Out', 'Stay In', 'Stay In', 'Go Out', 'Stay In']
}

# Convert categorical data to numeric
df = pd.DataFrame(data)
df['Weather'] = df['Weather'].map({'Sunny': 0, 'Rainy': 1})
df['Activity'] = df['Activity'].map({'Go Out': 0, 'Stay In': 1})

# Features and label
X = df[['Weather']]
y = df['Activity']

# Model
model = DecisionTreeClassifier(criterion='entropy', random_state=0)
model.fit(X, y)

# Plot the tree
plt.figure(figsize=(6, 4))
tree.plot_tree(model, feature_names=['Weather'], class_names=['Go Out', 'Stay In'], filled=True)
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn import tree

plt.figure(figsize=(6, 4))
tree.plot_tree(model,
               feature_names=['Weather'],       # only 1 feature
               class_names=['Go Out', 'Stay In'],  # based on how you encoded the labels
               filled=True,
               rounded=True,
               fontsize=12)
plt.show()
```

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
import matplotlib.pyplot as plt

# Create and encode data
data = {
    'Weather': ['Sunny', 'Sunny', 'Rainy', 'Rainy', 'Sunny', 'Rainy'],
    'Activity': ['Go Out', 'Go Out', 'Stay In', 'Stay In', 'Go Out', 'Stay In']
}
df = pd.DataFrame(data)
df['Weather'] = df['Weather'].map({'Sunny': 0, 'Rainy': 1})
df['Activity'] = df['Activity'].map({'Go Out': 0, 'Stay In': 1})

# Model training
X = df[['Weather']]
y = df['Activity']
model = DecisionTreeClassifier(criterion='entropy', random_state=0)
model.fit(X, y)

# Plot tree
plt.figure(figsize=(6, 4))
tree.plot_tree(model,
               feature_names=['Weather'],
               class_names=['Go Out', 'Stay In'],
               filled=True,
               rounded=True,
               fontsize=12)
plt.show()
```
